```python
maxScore=0
maxRS=0
```

```python
for i in range(1,200):
    dfx_train,dfx_test,dfy_train,dfy_test=train_test_split(dfx,dfy,test_size=.30,random_state=i)
    mod=LinearRegression()
    mod.fit(dfx_train,dfy_train)
    score=mod.score(dfx_train,dfy_train)
    mod.coef_
    pred=mod.predict(dfx_test)
    if score>maxScore:
        maxScore=score
        maxRS=i
print('Best score is ',maxScore,'on random state',maxRS)
```

```python
dfx_train,dfx_test,dfy_train,dfy_test=train_test_split(dfx,dfy,test_size=.30,random_state=107)
```
