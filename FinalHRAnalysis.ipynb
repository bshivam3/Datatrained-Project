Class imblanace issue is resolved

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
trainx=ss.fit_transform(trainx)
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import accuracy_score
```

```python
maxAccu=0
maxRS=0
```

```python
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(trainx,trainy,test_size=.30,random_state=i)
    model=DecisionTreeClassifier()
    model.fit(x_train,y_train)
    predm=model.predict(x_test)
    acc=accuracy_score(y_test,predm)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print('Best Accuracy is',maxAccu,'on Random State',maxRS)
```

```python
x_train,x_test,y_train,y_test=train_test_split(trainx,trainy,test_size=.30,random_state=94)
```

```python
DTC=DecisionTreeClassifier()
```

```python
DTC.fit(x_train,y_train)
```

```python
pred_dtc=DTC.predict(x_test)
```

```python
acc_dtc=classification_report(y_test,pred_dtc)
```

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
pred_gnb=gnb.predict(x_test)
acc_gnb=classification_report(y_test,pred_gnb)
print(acc_gnb)
```
