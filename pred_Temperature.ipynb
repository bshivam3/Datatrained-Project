```python
for i in range(1,201):
    x_train,x_test,y_train,y_test=train_test_split(x,ymax,test_size=.30,random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    pred=lr.predict(x_test)
    r2_s=r2_score(y_test,pred)
    if r2_s>max_r2:
        max_r2=r2_s
        max_rs=i
print('Best r2 score is ',max_r2,'on random state',max_rs)
```
